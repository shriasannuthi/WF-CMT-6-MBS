```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('ds2.csv')
data
```

```python
previous_UPB=[]

t1=data[0:1]

prev=float(t1['Original UPB'])
id=int(t1['Loan Identifier'])
```

```python
print(id)
```

```python
for i in range(len(data['Loan Identifier'])):
    
    if id==data.iloc[i,0]:
        previous_UPB.append(prev)
        prev=data.iloc[i,6]
    else:
        prev=data.iloc[i,6]
        previous_UPB.append(prev)
        id=data.iloc[i,0]
```

```python
data['Previous_UPB']=previous_UPB
```

```python
previous_UPB
```

```python
from datetime import datetime

def calculate_month_difference(date1, date2):
    date1 = datetime.strptime(date1, "%m%Y")
    date2 = datetime.strptime(date2, "%m%Y")

    months = (date2.year - date1.year) * 12 + (date2.month - date1.month)
    
    return months
```

```python
RMFM=[]

for i in range(len(data['Loan Identifier'])):
    RMFM.append(data.iloc[i,7]-calculate_month_difference(str(data.iloc[i,8]),str(data.iloc[i,1])))
```

```python
data['Remaining MFM']=RMFM
```
